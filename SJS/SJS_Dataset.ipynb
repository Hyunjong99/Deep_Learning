```python
def extract_numbers(input_string):
    return ''.join(char for char in input_string if char.isdigit())
```

```python
def extract_ID(filename):
    only_num = extract_numbers(filename)
    return only_num[:-4]
```

```python
root = f"/home/user1/Documents/Project2/AICOSS/Total_Data/"
save = f"/home/user1/Documents/Project2/AICOSS/PERSON/"

def get_IDLIST(root_path, save_path):
    for filename in os.listdir(root_path):
        if filename == "NORMAL" or filename == "NON_SJS" or filename == "SJS":
            diag_root = root_path + filename + f"/"
            diag_save = save_path + filename + f"/"
            for img in os.listdir(diag_root):
                    ID = extract_ID(img)
                    if ID not in os.listdir(diag_save):
                        os.makedirs(f"{diag_save}{ID}")
                    shutil.copy(f"{diag_root}{img}", f"{diag_save}{ID}/{img}")
```
